```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, expand, pprint

xs = [-1, 0, 1]
ys = [1, 5, 3]

a0, b0, c0, d0 = symbols("a0 b0 c0 d0")
a1, b1, c1, d1 = symbols("a1 b1 c1 d1")

eqs = [
    Eq(a0, ys[0]),                        # S0(-1) = 1
    Eq(a0 + b0 + c0 + d0, ys[1]),         # S0(0) = 5
    Eq(a1, ys[1]),                        # S1(0) = 5
    Eq(a1 + b1 + c1 + d1, ys[2]),         # S1(1) = 3
    Eq(b0 + 2*c0 + 3*d0, b1),             # S0'(0) = S1'(0)
    Eq(b0 + 2*c0 + 3*d0, -3),             # Pendiente deseada en x=0
    Eq(2*c0 + 6*d0, 2*c1),               
    Eq(2*c1 + 6*d1, 0),                   
]

sol = solve(eqs)

print("Coeficientes:")
for k, v in sol.items():
    print(f"{k} = {v}")

def Spline(x, x0, pars):
    a = float(pars["a"])
    b = float(pars["b"])
    c = float(pars["c"])
    d = float(pars["d"])
    return a + b*(x - x0) + c*(x - x0)**2 + d*(x - x0)**3
spline0 = { "a": sol[a0], "b": sol[b0], "c": sol[c0], "d": sol[d0] }
spline1 = { "a": sol[a1], "b": sol[b1], "c": sol[c1], "d": sol[d1] }

x0_vals = np.linspace(xs[0], xs[1], 100)
x1_vals = np.linspace(xs[1], xs[2], 100)
y0_vals = Spline(x0_vals, xs[0], spline0)
y1_vals = Spline(x1_vals, xs[1], spline1)

plt.plot(x0_vals, y0_vals, label="Spline 1", color="red")
plt.plot(x1_vals, y1_vals, label="Spline 2", color="blue")
plt.scatter(xs, ys, color="black", zorder=5)
plt.axvline(x=0, linestyle="--", color="gray", label="x=0")
plt.title("Splines cúbicos con pendiente m = -3 en x = 0")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

x = symbols("x")

print("\nExpresiones desarrolladas:")

S0_expr = sol[a0] + sol[b0]*(x + 1) + sol[c0]*(x + 1)**2 + sol[d0]*(x + 1)**3
print("\nS0(x) = ")
pprint(expand(S0_expr))

S1_expr = sol[a1] + sol[b1]*x + sol[c1]*x**2 + sol[d1]*x**3
print("\nS1(x) = ")
pprint(expand(S1_expr))

print(spline0)
print(spline1)
```
